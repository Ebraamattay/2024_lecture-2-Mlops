```python
import openpyxl
from openai import AzureOpenAI
import time

# Initialize AzureOpenAI client
client = AzureOpenAI(
    azure_endpoint="https://diabetes.openai.azure.com/openai/deployments/rrr/chat/completions?api-version=2024-02-15-preview",
    api_key='6f97cb14a59f4f0b91a8e05e31e90e33',
    api_version="2024-02-15-preview"
)

# Load Excel workbook
workbook = openpyxl.load_workbook('datagrad.xlsx')
sheet = workbook.active

# Initialize row counter
row_num = 4645  # Starting from row 103

# Retry parameters
max_retries = 3
retry_delay = 5  # seconds

# Iterate through each row and ask GPT-3
for row in sheet.iter_rows(min_row=row_num, max_row=min(row_num + 300, sheet.max_row), min_col=1, max_col=1, values_only=True):
    search_query = row[0]  # Assuming the search query is in the first column

    # Retry loop
    for retry_count in range(max_retries):
        try:
            # Ask GPT-3
            completion = client.chat.completions.create(
                model="rrr",  # model = "deployment_name"
                messages=[{"role": "system", "content": f"  احتاج الاجابة فقط نعم او لا دون اي شرح او توضيح هل هذه المعلومة صحيحة؟ {search_query}"}],
                temperature=0.7,
                max_tokens=800,
                top_p=0.95,
                frequency_penalty=0,
                presence_penalty=0,
                stop=None
            )

            # Process response
            response = completion.choices[0].message.content
            print(f"Row {row_num}: {response}")

            # Save the response in the fourth column
            sheet.cell(row=row_num, column=4, value=response)

            # Break out of the retry loop if successful
            break
        except Exception as e:
            print(f"Error occurred: {e}")
            if retry_count < max_retries - 1:
                print(f"Retrying in {retry_delay} seconds...")
                time.sleep(retry_delay)
            else:
                print("Max retries exceeded. Skipping this row.")
                break

    # Move to the next row
    row_num += 1

# Save the modified Excel file
workbook.save('your_updated_excel_file.xlsx')
```
